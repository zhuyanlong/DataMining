```python
#生成数据集
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(13)
#产生4个簇心5000行的随机数据
X,y=make_blobs(centers=4,n_samples=5000)
```

```python
plt.figure(figsize=(6,4))
#X[:,0]表示数据的第0列，X[:,1]表示数据的第1列,c用来指定颜色
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Dataset")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

```python
#类似于转置，把行数据转置为列数据
y=y.reshape((-1,1))
```

```python
#分割数据集，按比例划分
X_train,X_text,y_train,y_test=train_test_split(X,y)
```

```python
#将训练集字段和标签合在一起
train_dataset=np.append(X_train,y_train,axis=1)
```

```python
#将测试集字段和标签合在一起
test_dataset=np.append(X_text,y_test,axis=1)
```

```python
#保存结果
np.savetxt("train_dataset.txt", train_dataset, fmt="%.4f %.4f %d")
np.savetxt("test_dataset.txt", test_dataset, fmt="%.4f %.4f %d")
```
